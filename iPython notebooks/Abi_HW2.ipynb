```python
# import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load and access data
iris = pd.read_csv('data/iris_data.csv', names=['SepalLength',
                                                'SepalWidth',
                                                'PetalLength',
                                                'PetalWidth',
                                                'Name'])

```

```python
iris.info()
iris.head()
```

```python
# Create target variable
myfunc = lambda s: 0 if ((s['Name'] == 'Iris-setosa')) else 1 if ((s['Name'] == 'Iris-versicolor')) else 2
iris['Target'] = iris.apply(myfunc, axis=1)
```

```python
print iris
```

```python
# Set up dataset as arrays
columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
features = iris[list(columns)].values
features = numpy.delete(features, (-1), axis=0)
y = iris['Target'].values
y = numpy.delete(y, (-1), axis=0)
```

```python
# Import sklearn
from sklearn.cross_validation import train_test_split
```

```python
features_tr, features_te, y_tr, y_te = train_test_split( features, y, test_size=0.20, random_state=0)
```

```python
# Train KNN classifier defined function on the train data
from sklearn.neighbors import KNeighborsClassifier
myknn = KNeighborsClassifier(3).fit(features_tr,y_tr)
```

```python
# Question 1.
myknn.score(features_te, y_te)
```

```python
from sklearn.cross_validation import KFold
```

```python
# generic cross validation function
def cross_validate(features, y, classifier, k_fold) :

    # derive a set of (random) training and testing indices
    k_fold_indices = KFold( len(features), n_folds=k_fold,
                           indices=True, shuffle=True,
                           random_state=0)

    k_score_total = 0
    # for each training and testing slices run the classifier, and score the results
    for train_slice, test_slice in k_fold_indices :

        model = classifier(features[[ train_slice  ]],
                         y[[ train_slice  ]])

        k_score = model.score(features[[ test_slice ]],
                              y[[ test_slice ]])

        k_score_total += k_score

    # return the average accuracy
    return k_score_total/k_fold
```

```python
# Question 2
cross_validate(features, y, KNeighborsClassifier(3).fit, 5)
```

```python
# Question 3
results = []
for n in range(1, 150, 1):
    check = KNeighborsClassifier(n_neighbors=n)
    check.fit(features_tr, y_tr)
    preds = check.predict(features_te)
    accuracy = np.where(preds==y_te, 1, 0).sum() / float(len(y_te))
    print "Neighbors: %d, Accuracy: %3f" % (n, accuracy)
 
    results.append([n, accuracy])
 
results = pd.DataFrame(results, columns=["n", "accuracy"])
```

```python
# It looks like a K = 1 would be the optimal value to use, the next closest being 4
```

```python
# Question 4
plt.plot(results.n, results.accuracy)
plt.title("Effect with increasing K")
plt.show()
```
